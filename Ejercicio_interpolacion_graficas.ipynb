```python
import numpy
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
x1=numpy.arange(-5,5,0.1)
y1=(5*x1**3)-(0.1*(x1**4))+1.2

aleat = numpy.random.rand(5)
x2=5-(10*aleat)
x2=numpy.append([x2],[-5])
x2=numpy.append([x2],[5])

y2=(5*x2**3)-(0.1*(x2**4))+1.2

fun=interpolate.interp1d(x2, y2)
y3=fun(x1)



lin1=plt.plot(x1,y1,label="Original")
lin2=plt.plot(x1,y3,label="Aprox. con interpolacion")
pts1=plt.plot(x2,y2,'o',label="Muestras")
plt.grid()
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
plt.savefig('graf.png')
plt.show()

```
